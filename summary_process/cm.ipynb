```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#  Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20

# 
# 
true_labels = ["normal", "dos", "probe", "r2l", "u2r"]
predicted_labels = ["normal", "dos", "probe", "r2l", "u2r"]

# Thirty-Fifth Matrix
confusion_data = np.array([
    [18359, 123, 190, 591, 1], 
    [4, 13340, 3, 0, 0], 
    [13, 1, 3504, 1, 0], 
    [21, 1, 6, 939, 3], 
    [5, 0, 0, 9, 16]
])
# 
plt.figure(1, figsize=(10, 8))
sns.set(font_scale=2)
sns.set(color_codes=True)
plt.title('Confusion Matrix', fontsize=20,fontname='Times New Roman')

# 
ax = sns.heatmap(confusion_data, annot=True, fmt="d", cmap="Blues",
                 xticklabels=predicted_labels, yticklabels=true_labels,annot_kws={"size": 20})

# 
ax.set_xlabel('Predicted Label', fontsize=20,fontname='Times New Roman')
ax.set_ylabel('True Label', fontsize=20,fontname='Times New Roman')

# 
ax.tick_params(axis='both', which='major', labelsize=20)

#  Times New Roman
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontname('Times New Roman')
    label.set_fontsize(20)

plt.show()
```
