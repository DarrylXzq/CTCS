```python
import pandas as pd
import matplotlib.pyplot as plt
from typing import Dict
```

```python
plt.rcParams['font.size'] = 9
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'


def plot_model_performance(csv_files: Dict[str, str]):
    plt.figure(figsize=(10, 4))
    for model_name, csv_file in csv_files.items():
        data = pd.read_csv(csv_file)
        plt.plot(data['val_loss'], label=f'{model_name}')
    # plt.title('Inception Validation Loss Over Epochs', fontweight='normal')
    plt.xlabel('Epoch', fontweight='normal')
    plt.ylabel('Validation Loss', fontweight='normal')
    plt.legend()
    # plt.ylim(0, 0.7)
    # plt.ylim(0, 0.5)
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 4))
    for model_name, csv_file in csv_files.items():
        data = pd.read_csv(csv_file)
        plt.plot(data['val_accuracy'], label=f'{model_name}')
    # plt.title('Inception Validation Accuracy Over Epochs', fontweight='normal')
    plt.xlabel('Epoch', fontweight='normal')
    plt.ylabel('Validation Accuracy', fontweight='normal')
    plt.legend()
    # plt.ylim(0.2, 1.02)
    # plt.ylim(0.5, 1.02)
    plt.grid(True)
    plt.show()
```
